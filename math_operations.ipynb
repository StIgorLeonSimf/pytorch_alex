```python
# shape(4,3,5)  и число 55 всегда согасованы
# shape(4,3,5)  и масив [1] shape(1,) с одним элементом всегда согасованы
# shape(4,3,5) и  shape(5,) или сами или автоматически (1,1,5)
# shape(4,3,5) и  shape(4,) -> expand_dims(arr, axis=[1, 2]) (4, 1, 1)
# shape(4,3,5) и  shape(4, 5) -> arr[:,np.newaxis]  ( 4, 1, 5)
```
