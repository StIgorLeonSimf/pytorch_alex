```python
%%time
data = np.empty([1000, 28, 28])
for i in range(1000):
    img = np.random.randint(0, 256, (28, 28))
    data = np.append(data, img[np.newaxis], axis=0)
print(data.shape)
```

```python
%%time
data = np.empty([1000, 28, 28])
data.resize([2000, 28, 28], refcheck=False)
for i in range(1000):
    img = np.random.randint(0, 256, (28, 28))
    data[1000 + i] = img
print(data.shape)
```
