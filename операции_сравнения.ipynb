```python
# скалярное произведение векторов
np.random.seed(42)
vector1 = np.array([1, 2, 3, 4, 5])
vector2 = np.array([2, 3, 4, 5, 6])
matrix1 = np.random.randint(10, size=[3, 3])
matrix2 = np.random.randint(10, size=[3, 3])
# vector2.shape = [-1, 1]
# vector2
np.dot(vector1, vector2)
vector1 @ vector2
np.inner(vector1, vector2)
#Внешнее произведение векторов
np.outer(vector1, vector2)
```

```python
#умножение матриц
vec1 = np.array([1, 2, 3])
vec2 = np.array([2, 3, 4, 5])
mat = np.random.randint(10, size=[4, 3])
print(matrix1)
print(matrix2)
np.dot(matrix1, matrix2)
matrix1 @ matrix2
np.matmul(matrix1, matrix2)
```

```python
# Матрица на вектор
print(vec2)
print(mat)

print(np.dot(vec2, mat))
print(mat @ vec1)
print(mat @ vec1[:np.newaxis])
```

```python
# элементы линейной алгебры
np.linalg.matrix_rank(mat) # ранг матрицы
np.linalg.inv(mat[:3]) # обратная матрица
np.trace(mat)  # cлед матрицы. Сумма на главной диагонали
np.linalg.det(mat[:3]) # определитель матрицы
np.linalg.eig(mat[:3]) # собственые значения и собственные вектора матрицы
np.linalg.svd(mat) # сингулярное разложение
```
